```python
# Cell 1: Setup and Imports
import sys
import os

# Add the agents directory to Python path
sys.path.append('.')
sys.path.append('../simulation')


from party_agent import PartyAgent
from party_discussion import discuss_legislation

# Cell 2: Create Parties and Add Politicians
print("=== TWORZENIE PARTII POLITYCZNYCH ===\n")

# Koalicja Obywatelska (Civic Coalition)
ko = PartyAgent(name="Koalicja Obywatelska", acronym="KO")
ko.add_politician("Donald Tusk")
ko.add_politician("Rafał Trzaskowski")
ko.add_politician("Barbara Nowacka")
ko.add_politician("Borys Budka")

# Prawo i Sprawiedliwość (Law and Justice)
pis = PartyAgent(name="Prawo i Sprawiedliwość", acronym="PiS")
pis.add_politician("Jarosław Kaczyński")
pis.add_politician("Mateusz Morawiecki")
pis.add_politician("Beata Szydło")
pis.add_politician("Mariusz Błaszczak")

# Konfederacja (Confederation)
konfederacja = PartyAgent(name="Konfederacja", acronym="Konfederacja")
konfederacja.add_politician("Sławomir Mentzen")
konfederacja.add_politician("Krzysztof Bosak")
konfederacja.add_politician("Konrad Berkowicz")

print("\n✓ Wszystkie partie utworzone!")

# Cell 3: Define Legislation to Discuss
legislation = """
Ustawa o wprowadzeniu 32-godzinnego tygodnia pracy:
- Skrócenie tygodnia pracy z 40 do 32 godzin
- Zachowanie pełnego wynagrodzenia
- Wejście w życie od 2025 roku
- Możliwość elastycznego rozłożenia godzin pracy
"""

print(f"\n=== USTAWA DO DYSKUSJI ===")
print(legislation)

# Cell 4: Run Internal Party Discussions
print("\n=== ROZPOCZYNAMY DYSKUSJE WEWNĄTRZPARTYJNE ===")

parties = [ko, pis, konfederacja]
positions = discuss_legislation(parties, legislation)

# Cell 5: Display Results Summary
print("\n" + "="*80)
print("PODSUMOWANIE STANOWISK PARTII")
print("="*80)

for party_name, position in positions.items():
    print(f"\n{party_name}:")
    print(f"  Stanowisko: {'✓ POPIERA' if position.supports_legislation else '✗ PRZECIW'}")
    print(f"  Główne argumenty:")
    for i, arg in enumerate(position.main_arguments, 1):
        print(f"    {i}. {arg}")
    
    print(f"\n  Głosy indywidualne:")
    for politician, stance in position.individual_opinions.items():
        print(f"    • {politician}: {stance}")

# Cell 6: Voting Summary
print("\n" + "="*80)
print("SYMULACJA GŁOSOWANIA")
print("="*80)

votes_for = 0
votes_against = 0

for party_name, position in positions.items():
    party = next(p for p in parties if p.name == party_name)
    party_votes = len(party.politicians)
    
    if position.supports_legislation:
        votes_for += party_votes
        print(f"{party_name}: {party_votes} głosów ZA")
    else:
        votes_against += party_votes
        print(f"{party_name}: {party_votes} głosów PRZECIW")

print(f"\nŁącznie głosów ZA: {votes_for}")
print(f"Łącznie głosów PRZECIW: {votes_against}")
print(f"\nWYNIK: Ustawa {'PRZYJĘTA ✓' if votes_for > votes_against else 'ODRZUCONA ✗'}")

# Cell 7: Test Different Legislation (Optional)
# Uncomment to test with different legislation
"""
print("\n\n=== TESTUJEMY INNĄ USTAWĘ ===")

legislation2 = '''
Ustawa o podatku od majątku:
- Wprowadzenie 2% podatku od majątku powyżej 10 mln PLN
- Zwolnienie z podatku pierwszego mieszkania
- Środki przeznaczone na służbę zdrowia
'''

positions2 = discuss_legislation(parties, legislation2)

for party_name, position in positions2.items():
    print(f"\n{party_name}: {'POPIERA' if position.supports_legislation else 'PRZECIW'}")
"""
```
